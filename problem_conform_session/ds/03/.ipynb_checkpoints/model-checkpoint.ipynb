```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pickle
```

```python
model = LinearRegression()
```

```python
df_train = pd.read_csv("train.csv")
df_train.drop(['id', 'store_id', 'day'], axis='columns', inplace=True)
df_train
```

```python
df_train.isnull().sum()
```

```python
df_train.info()
```

```python
df_train.dropna(inplace=True)
df_train.info()
```

```python
df_weather = pd.get_dummies(df_train, columns=['weather'], dtype=int)
df_season = pd.get_dummies(df_weather, columns=['season'], dtype=int)
df_season
```

```python
df_train = df_season
```

```python
df_train
```

```python
a = list(df_train.columns)
a.remove('sales')
print(a)
X_train = df_train[a]
y_train = df_train['sales']
```

```python
model.fit(X_train, y_train)
print(model.intercept_, model.coef_)
```

```python
model_filename = 'text_classification_model.pkl'
with open(model_filename, 'wb') as file:
    pickle.dump(model, file)

print(f"モデルは{model_filename}に保存されました。")
```
